# EDA

# Chances of success to match in a speed dating

```python
# Creation of a column that define if there is a match or no
dataset['success'] = dataset.match.apply(lambda x: 'Yes' if x > 0 else 'No')
```

```python
# Defining the ratio of success
success_ratio = (dataset['success'].value_counts(normalize=True)*100).rename_axis('success').reset_index(name='counts')
```

```python
# Showing success proportion
fig1 = px.pie(success_ratio,
             values='counts',
             names='success', 
             color ='success',
             title='Proportion of success',
             width= 1000,
             color_discrete_map={'Yes':'Blue',
                                 'No':'Orange'}
             )
fig1.update_traces(textposition = 'outside', textfont_size = 15)             
fig1.update_layout(title_x = 0.5, 
                  margin=dict(l=50,r=50,b=50,t=50,pad=4),
                  )                
fig1.show()
```

We can see that the success ratio to have a match in a speed dating event is very low.

# Axis 1: What is the impact of age in finding love ?

```python
# Age of participants by gender
fig2 = px.histogram(dataset, 
                    x="age",
                    color="gender",
                    title = "Age of participants by gender",
                    color_discrete_map = {1:'blue', 0:'red'},
                    width=1500, height=500
                    )
fig2.update_layout(autosize = True,
                  title_x = 0.5,
                  width=800,
                  height=500,
                  margin=dict(l=50,r=50,b=50,t=50,pad=4)
                  )
fig2.show()
```

We can see that the vast majority of the participants are between 21 and 30. The distribution between gender is fairly distributed equally. Seems that the primary target for speed dating are young people.

```python
# Match average by age
fig3 = px.histogram(dataset, x="age",
                   y="match",
                   color= 'gender',
                   title = 'Match average by age',
                   barmode ='group',
                   histfunc= 'avg',
                   )
fig3.update_traces(textposition = 'outside', textfont_size = 15)                  
fig3.update_layout(autosize = True,
                  title_x = 0.5,
                  width=800,
                  height=500,
                  margin=dict(l=50,r=50,b=50,t=50,pad=4)
                  )
fig3.update_xaxes(tickfont_size=15,tickmode='linear')                  
fig3.show()
```

It's interesting to see that during the majority of the range (21-30), both mens and womens are fairly equal in terms of matchs. Before the age of 25, womens matchs more than mens. The peak for both gender is at 25, mid twenties seems to be a good age for having matchs. Late twenties are still favorable to womens but if we look outside the range where the majority of participants belong we can see that mens are taking the advantage. We can suppose that age is more of an advantage for mens after 30 and age is more of an advantage for womens in their twenties.

```python
# Creation of a new column to calculate the age difference
dataset['age_difference'] = abs(dataset['age_o']- dataset['age'])
```

```python
# Age difference between participants
fig4 = px.histogram(dataset, 
             x="age_difference",
             title= "Age difference between participants"
             )

fig4.update_layout(title_x = 0.5,
                  autosize = True,
                  width=800,
                  height=500,
                  margin=dict(l=50,r=50,b=50,t=50,pad=4),
                  yaxis_title = 'age',
                  yaxis = {'visible': False},
                  xaxis_title = 'age difference',
                  )                              
fig4.show()
```

We can clearly see here that the age difference is very close between participants with the majority having a difference from 0 to 6 years. Let's see how it impact the matchs.

```python
# Display match average by age difference
fig5 = px.histogram(dataset, x="age_difference",
                   y="match",
                   title = 'Average match by age difference',
                   histfunc= 'avg'
                   )
fig5.update_traces(textposition = 'outside', textfont_size = 15)                  
fig5.update_layout(autosize = True,
                  title_x = 0.5,
                  width=800,
                  height=500,
                  margin=dict(l=50,r=50,b=50,t=50,pad=4),
                  xaxis_title = 'age difference',
                  )
fig5.update_xaxes(tickfont_size=15, tickmode='linear')                  
fig5.show()
```

The average match regarding age difference is pretty stable even if it decreases little by little until 10 years of difference before to drop completely and surprisingly going up again at 13 years of difference. Surprinsingly the average match is the higher of all with a age difference of 17 but we cannot be sure that it is very significant in this case.

# Axis 2: What is the impact of mutual interests and subjectives criterias in finding love ?

```python
# Keeping only the interests
interest_corr = dataset[dataset.columns[dataset.columns.isin(["sports", "tvsports", "exercise", "dining", "museums", "art", "hiking", "gaming", "clubbing", "reading", "tv", "theater", 
"movies", "concerts", "music", "shopping", "yoga"])]]
```

```python
# Correlation heatmap between interests
corr1 = interest_corr.corr()
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr1, ax=ax)
```

We can see that the interests in similar arts are very correlated.
Museums, art and movies with theater and concerts with music.

```python
# Keeping only the subjectives criterias
criterias_corr = dataset[dataset.columns[dataset.columns.isin(["dec", "attr", "sinc", "intel", "fun", "amb", "shar", "like", "prob"])]]
```

```python
# Correlation heatmap between interests
corr2 = criterias_corr.corr()
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr2, ax=ax)
```

We can see that the fact of liking someone is very correlated to the attractivity and the fun of the person.
Intelligence is very correlated too sincerity and ambition.

```python
# Printing the correlation score for the decision factors
corr2["dec"].apply(abs).sort_values(ascending=False)
```

With no surprise the fact of liking someone is at the top of the decision factors of matching with someone followed by attractiveness and fun. Intelligence, sincerity and ambitions are at the bottom of the decision factors.
In a very quick speed dating it's impossible to determine profound and core values of a person and much easier to take a decision based on superficial criterias.

# Conclusion

It's very hard to be successful in a speed dating with only 16% chances of getting a match.

Regarding the impact of age, both mens and womens have equal chances to get a match if they are in their twenties. The chances are higher for womens in their early twenties when chances are higher for mens when they passed the 30's.
The age difference seems to have an impact too with chances decreasing the higher the age difference is even though there is a big average match with a difference of 17 years, we can't take it as a generality but more located to this specific waves of events.

Regarding mutual interests now, there is strong correlation with interests in the same area such as theater, art and museums for example. Finally regarding subjective criterias, we have seen that the fact of liking a person is highly correlated to the attractiveness and fun of this person. It's no surprise to see "superficial" criterias with the highest correlation scores meanwhile profound and core values are at the bottom, it's time related to the dates rounds that last only for a few minutes.
